1.What is an API?

 An API (Application Programming Interface) is a set of rules that allows different software applications to communicate and share data with each other.

 2.What is an AI API?

  AI API is an interface that allows applications to use artificial intelligence features, such as text generation, image creation, or speech recognition.

  3.What is Gemini API?

   Gemini API is an AI API developed by Google that allows developers to integrate Gemini AI models into their applications for tasks like text generation, image understanding, coding assistance, and question answering.

   4.Difference between normal Python program and AI API chatbot

1.Normal Python Program follows fixed rules written by the programmer.

2.AI API Chatbot uses an AI model to generate intelligent responses.

3.Normal Python Program gives predefined output for a specific input.

4.AI API Chatbot provides dynamic and context-aware answers.

5.Normal Python Program cannot understand natural language deeply.

6.AI API Chatbot can understand and respond to natural language.

7.Normal Python Program works only with coded logic.

8.AI API Chatbot can answer a wide variety of questions.

9.Normal Python Program runs on local code without AI.

10.AI API Chatbot communicates with an AI service through an API.

Short summary:

Normal Python Program = Fixed logic

AI API Chatbot = AI-powered intelligent response

 6.Create a simple 5-line AI chatbot in Python     

```python
from google import genai
client = genai.Client(api_key="api_key")
while True:
    user = input("You: ")
    print("Bot:", client.models.generate_content(model="gemini-2.5-flash", contents=user).text)
```
