1.What is a tokenizer?

* A tokenizer is a tool that breaks text into smaller units called tokens.
* Tokens can be words, subwords, or characters.
* It converts text into a format that a language model can understand.
* Each token is assigned a unique numerical ID.
* Tokenization is the first step in processing text for AI models.

2.Why is tokenization needed?

* It breaks text into smaller tokens for processing.
* It converts text into numerical IDs that language models can understand.
* It helps the model process text efficiently.
* It enables the model to learn patterns and relationships between words.
* It supports handling of large vocabularies by splitting words into subwords when needed.
* It is the first step before creating embeddings and making predictions

3.Types of tokenizers.

* Word Tokenizer: Splits text into individual words.

Example: "I love AI" → ["I", "love", "AI"]
* Character Tokenizer: Splits text into individual characters.

Example: "AI" → ["A", "I"]
* Subword Tokenizer: Splits words into smaller meaningful parts.

Example: "playing" → ["play", "##ing"]
* Sentence Tokenizer: Splits text into separate sentences.

Example: "Hello. How are you?" → ["Hello.", "How are you?"]
* Byte Pair Encoding (BPE): Uses frequently occurring subword units to handle unknown words efficiently.
* WordPiece Tokenizer: Breaks words into subwords and is widely used in BERT models.
* SentencePiece Tokenizer: Works directly on raw text without requiring spaces and is commonly used in multilingual models.

4.How tokenization impacts model performance.

* Improves text processing by converting text into manageable tokens.
* Reduces vocabulary size using subword tokenization, making models more efficient.
* Handles unknown or rare words by breaking them into smaller parts.
* Improves accuracy by preserving meaningful word patterns.
* Affects memory and speed—fewer tokens generally lead to faster processing.
* Enhances language understanding by creating consistent input for the model.
* Influences training and inference quality, as better tokenization leads to better model performance.
