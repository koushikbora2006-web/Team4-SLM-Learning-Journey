1.What are Transformers?

A Transformer is a neural network architecture that processes text efficiently by understanding relationships between words using self-attention, enabling SLMs to perform tasks like text generation, translation, and question answering

2.Why were they introduced?

Transformers were introduced to process text faster, capture long-range dependencies better, and allow parallel processing, leading to improved performance in NLP tasks such as translation, summarization, and chatbots.

3.Basic architecture overview.

A Transformer mainly consists of two parts:

Encoder
* Reads the input text.
* Understands the meaning and context of the words.

Decoder
* Generates the output text one word at a time.
* Uses the encoder's information to produce the final response.

Main Components
* Input Embedding: Converts words into numerical vectors.
* Positional Encoding: Adds word order information.
* Self-Attention: Determines which words are important to each other.
* Feed-Forward Network: Processes the attention output.
* Output Layer: Predicts the next word or token.

Simple Flow

Input Text → Embedding → Positional Encoding → Self-Attention → Feed-Forward Network → Output

4.Role of Attention Mechanism.

Role of the Attention Mechanism

The Attention Mechanism helps a Transformer focus on the most relevant words in a sentence while processing or generating text.

Key Roles:
* Identifies important words related to the current word.
* Captures long-range relationships between words in a sentence.
* Improves understanding of context and meaning.
* Enables parallel processing, making the model faster than RNNs.

Example:

In the sentence "The cat chased the mouse because it was hungry," the attention mechanism helps the model understand that "it" refers to "the cat."
