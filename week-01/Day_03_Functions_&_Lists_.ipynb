Create a Multiplication Table Generator using Functions

```python
# Function to generate multiplication table
def multiplication_table(num):
    print(f"\nMultiplication Table of {num}\n")

    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

# Main program
number = int(input("Enter a number: "))

# Function call
multiplication_table(number)
```

Create a Simple Data Store using Lists

```python
# Create an empty list
data_store = []

# Get the number of items to store
n = int(input("How many items do you want to store? "))

# Store items in the list
for i in range(n):
    item = input(f"Enter item {i + 1}: ")
    data_store.append(item)

# Display the stored items
print("\nStored Data:")
for item in data_store:
    print(item)
```

Add Student Names into a List

```python
# Create an empty list
students = []

# Input the number of students
n = int(input("Enter the number of students: "))

# Add student names to the list
for i in range(n):
    name = input(f"Enter name of student {i + 1}: ")
    students.append(name)

# Display the student names
print("\nList of Student Names:")
for name in students:
    print(name)
```

Display all Student Names using a Loop

```python
# List of student names
students = ["Raj", "Lara", "Wong"]

# Display all student names using a loop
print("Student Names:")

for student in students:
    print(student)
```

Create a Function to Add Numbers

```python
# Function to add two numbers
def add_numbers(a, b):
    return a + b

# Input from the user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Function call
result = add_numbers(num1, num2)

# Display the result
print("Sum =", result)
```

Create a Function to Display Student Details

```python
# Function to display student details
def display_student(name, age, course):
    print("\nStudent Details")
    print("Name   :", name)
    print("Age    :", age)
    print("Course :", course)

# Input from the user
name = input("Enter student name: ")
age = int(input("Enter student age: "))
course = input("Enter student course: ")

# Function call
display_student(name, age, course)
```
