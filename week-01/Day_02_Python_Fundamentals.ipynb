4. Positive, Negative, or Zero

Problem Statement:
Take a number as input and determine whether it is:

● Positive

● Negative

● Zero

```python
num = int(input("Enter a number:"))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

5. Largest of Two Numbers

Problem Statement:
Take two numbers as input and find the larger number.

```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))

if num1 > num2:
    print(num1, "is larger")
elif num2 > num1:
    print(num2, "is larger")
else:
    print("Both numbers are equal")
```

8. Multiplication Table

Problem Statement:
Take a number as input and generate its multiplication table.

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(f"{num}{i} = {num * i}")
```

9. Sum of Numbers from 1 to N

Problem Statement:
Take a number N and find the sum from 1 to N.

```python
n = int(input("Enter a number: "))
sum = 0
for i in range(1, n + 1):
    sum = sum + i
print("Sum =", sum)
```

10. Count Even and Odd Numbers

Problem Statement:
Take a number N and count:

● Total Even Numbers

● Total Odd Numbers
between 1 and N.

```python
n = int(input("Enter a number: "))
even = 0
odd = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        even += 1
    else:
        odd += 1
print("Even Count =", even)
print("Odd Count =", odd)
```
